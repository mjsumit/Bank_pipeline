```python
## Final customers ----- 

import pandas as pd
import random

fname = ["sumit", "pratik", "suresh", "irfan", "Rithun", "Sooraj", "farok", "raajil", "akhil", "rakshi"]
lname = ["mj", "br", "S", "C", "cs", "tm", "cp", "ts", "J", "muhamad"]

cust_id = []
cust_name = []
acc_type = []
risk_score = []

for i in range(1000):
    unique_id = 1001 + i
    cust_id.append(unique_id)

    # FIXED: Added the 'f', corrected the spelling to 'choice', and added a space between names
    random_name = f"{random.choice(fname)} {random.choice(lname)}"
    
    # FIXED: Kept only this name logic block, removed the old generic "Customer_1001" block
    if i % 10 == 0:
        cust_name.append(random_name.lower())
    else:
        cust_name.append(random_name)
        
    acc_type.append(random.choice(["savings", "current", "fixed"]))
    risk_score.append(random.randint(1, 5))

# Pack into dictionary
lcustomers = {
    "CUST_ID": cust_id,
    "CUST_NAME": cust_name,
    "ACC_TYPE": acc_type,
    "RISK_SCORE": risk_score
}

# Convert and display
#lcustomers = pd.DataFrame(lcustomers)
 # Displaying 12 rows so you can see the 0th and 10th row anomaly!
```

```python
lcustomers = pd.DataFrame(lcustomers)
lcustomers
```

```python
lcustomers.to_csv("lcustomers_raw.csv",index=False)
```

```python
# final Transactions----


txn_id =[]
cust_id_txn =[]
txn_type = []
txn_amount = []
txn_dates = []

for i in range (5000): 
    txn_id.append(f"TXN_{2001 + i}")

    cust_id_txn.append(random.randint(1001,1100))

    txn_type.append(random.choice(["deposit","withdrawl","transfer"]))

    txn_amount.append(random.randint(1000,55000))

    txn_dates.append(f"2026-06-{random.randint(1,17)}")

ltransactions={
    "TXN_ID" : txn_id,
    "CUST_ID" : cust_id_txn,
    "TXN_TYPE" : txn_type,
    "TXN_AMOUNT" : txn_amount,
    "TXN_DATES" : txn_dates
}
```
